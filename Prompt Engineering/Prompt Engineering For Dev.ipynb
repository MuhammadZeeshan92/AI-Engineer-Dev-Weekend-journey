````python
# # Prompt Engineering Techniques

# This notebook demonstrates various prompt engineering techniques learned from the DeepLearning.AI "ChatGPT Prompt Engineering for Developers" course.

# The notebook covers:

# - Zero-shot prompting
# - Few-shot prompting
# - System/User/Assistant roles
# - Delimiters
# - Output formatting
# - Step-by-step prompting
# - Hallucination mitigation

# All examples use the OpenAI Chat Completion API.
````

````python
from openai import OpenAI

client = OpenAI(
    api_key="sk-or-v1....................",
    base_url="https://openrouter.ai/api/v1",
)
````

````python
def get_completion(messages, temperature=0):
    response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=messages,
        temperature=temperature
    )
    return response.choices[0].message.content
````

````python
# ## Example 1 — Zero-Shot Prompting

# The model performs a task without seeing any examples.
````

````python
messages = [
    {
        "role":"user",
        "content":"Classify the sentiment of this review. Reply only with Positive or Negative.\n\nThe laptop is amazing and the battery lasts all day."
    }
]

print(get_completion(messages))
````

````python
# Zero-shot prompting gives only instructions.
# The model already knows how to perform sentiment analysis without training examples.
````

````python
# ## Example 2 — Few-Shot Prompting

# Provide examples before asking the real question.
````

````python
messages = [
{
"role":"user",
"content":"""
Translate English to French.

Dog -> Chien
Cat -> Chat
House -> Maison

Car ->
"""
}
]

print(get_completion(messages))
````

````python
# The examples teach the desired pattern before asking the final question.
````

````python
# Example 3 — System, User and Assistant Roles
````

````python
messages = [

{
"role":"system",
"content":"You are a helpful travel assistant."
},

{
"role":"user",
"content":"Suggest two places to visit in Pakistan."
}

]

print(get_completion(messages))
````

````python
# The system message defines the assistant's behavior.

# The user message contains the request.

# The assistant generates the response.
````

````python
# Example 4 — Delimiters and JSON Output
````

````python
review = """
The phone arrived early.
Battery life is excellent.
Camera quality is average.
"""

messages = [

{
"role":"user",
"content":f"""
Analyze the review below.

Review:
```{review}```

Return JSON with:

sentiment

pros

cons
"""
}

]

print(get_completion(messages))
````

````python
# Delimiters clearly separate instructions from data.

# JSON makes the output easier for software systems.
````

````python
## Example 5 — Step-by-Step Prompting
````

````python
article = """
Artificial Intelligence is changing healthcare by improving diagnosis,
predicting diseases and assisting doctors.
"""

messages = [

{
"role":"user",
"content":f"""
Read the article.

Step 1:
Identify the main topic.

Step 2:
List two important facts.

Step 3:
Write a one-sentence summary.

Article:
```{article}```
"""
}

]

print(get_completion(messages))
````

````python
# Instead of asking the model to do everything at once,
# the task is divided into smaller reasoning steps.

# This follows the "Give the model time to think" principle.
````

````python
# # Hallucination

# Hallucination occurs when a Large Language Model generates information that sounds correct but is actually false or unsupported.

# Example:
# If asked about a product with missing information, the model might invent specifications.

# Prompt engineering can reduce hallucinations by:

# - Using clear instructions
# - Providing sufficient context
# - Using delimiters
# - Asking the model to answer "unknown" when information is unavailable
# - Requesting structured outputs such as JSON
````

````python
# ==========================================================
# Combined Prompt Engineering Example
# Demonstrates multiple prompt engineering techniques together
# ==========================================================

review = """
I ordered a wireless keyboard. It arrived one day earlier than expected.
The typing experience is excellent and the battery lasts a very long time.
However, one key occasionally sticks. Overall, I am satisfied with the purchase.
"""

messages = [

    # System Role
    {
        "role": "system",
        "content": """
You are a professional customer support assistant.
Always provide clear, accurate and helpful responses.
If information is missing, reply with "unknown" instead of guessing.
"""
    },

    # User Prompt
    {
        "role": "user",
        "content": f"""
Analyze the customer review delimited by triple backticks.

Review:
```{review}```

Perform the following tasks step by step.

Step 1:
Determine the sentiment.

Step 2:
Identify the product.

Step 3:
Extract two pros and one con.

Step 4:
Write a short customer service reply.

Step 5:
Return everything as JSON using the following structure.

{{
    "sentiment":"",
    "product":"",
    "pros":[],
    "cons":[],
    "reply":""
}}
"""
    }

]

print(get_completion(messages))
````

````python
# # Prompt Engineering for Developers — Complete Course Summary

# ## Introduction

# Prompt Engineering is the process of designing clear, precise, and well-structured instructions (prompts) that guide a Large Language Model (LLM) to produce the desired output. Unlike traditional machine learning, where a separate model must be trained for every new task, prompt engineering allows a single pretrained LLM to perform thousands of different tasks simply by changing the prompt. This makes LLMs extremely flexible and powerful for real-world applications.

# The main goal of prompt engineering is not just to ask a question but to communicate effectively with the model so that it understands exactly what is expected. The better the prompt, the better the output.

# ---

# # Traditional Machine Learning vs Prompt Engineering

# Traditional Machine Learning follows a long development pipeline:

# Collect Data
# → Label Data
# → Train Model
# → Evaluate
# → Deploy
# → Inference

# Every new task (such as spam detection, sentiment analysis, or topic classification) often requires collecting new data and training a new model.

# Prompt Engineering changes this workflow completely.

# Prompt
# → Large Language Model
# → Output

# Instead of training multiple models, we use one powerful language model and change only the prompt according to the task.

# For example, the same model can:

# - Translate languages
# - Summarize articles
# - Detect sentiment
# - Generate emails
# - Extract information
# - Answer questions
# - Write code
# - Build chatbots

# without any additional training.

# ---

# # Two Fundamental Principles of Prompt Engineering

# The course introduces two core principles that form the foundation of effective prompting.

# ## 1. Write Clear and Specific Instructions

# The clearer the prompt, the better the output.

# Poor Prompt:

# Summarize this article.

# Better Prompt:

# Summarize this article in fewer than 30 words.
# Focus only on shipping-related information.
# Return the result as bullet points.

# Being specific removes ambiguity and helps the model understand exactly what is expected.

# Good prompts often specify:

# - Task
# - Audience
# - Output format
# - Length
# - Constraints
# - Style

# ---

# ## 2. Give the Model Time to Think

# Many complex problems become easier when broken into smaller steps.

# Instead of asking:

# Analyze this article.

# Use:

# Step 1:
# Identify the important facts.

# Step 2:
# Determine the sentiment.

# Step 3:
# Generate a summary.

# Breaking large problems into smaller subtasks significantly improves reasoning quality and accuracy.

# ---

# # Best Prompt Engineering Practices

# Throughout the course, several best practices are repeatedly emphasized.

# ## Use Delimiters

# Delimiters clearly separate instructions from input data.

# Common delimiters:

# ```
# """
# '''
# ```

# or

# ```
# ```
# ```

# Example:

# Summarize the text below.

# Text:
# ```{article}```

# Benefits:

# - Separates instructions from data
# - Prevents confusion
# - Makes prompts easier to understand
# - Reduces accidental interpretation of input as instructions

# ---

# ## Request Structured Output

# Instead of asking for free-form paragraphs, request structured formats such as:

# - JSON
# - HTML
# - Markdown
# - Tables
# - Lists

# Example:

# Return the answer as JSON.

# Example output:

# ```json
# {
#     "sentiment": "positive",
#     "brand": "Apple"
# }
# ```

# Structured outputs are much easier for software systems to process.

# ---

# ## Check Conditions

# Never assume information exists.

# Instead, explicitly instruct the model:

# "If the information is unavailable, return 'unknown'."

# This reduces hallucinations and improves reliability.

# ---

# ## Few-Shot Prompting

# Instead of giving only instructions, provide examples.

# Example:

# Input:
# Hello

# Output:
# Bonjour

# Input:
# Dog

# Output:
# Chien

# Input:
# House

# The model continues the learned pattern.

# Few-shot prompting is useful when the desired format is difficult to describe using words alone.

# ---

# # Temperature

# Temperature controls randomness in the model's responses.

# Low Temperature (0)

# Characteristics:

# - Deterministic
# - Predictable
# - Repeatable
# - More factual

# Best for:

# - Classification
# - Extraction
# - JSON generation
# - Customer support
# - Production systems

# High Temperature (0.7–1.0)

# Characteristics:

# - More creative
# - More varied
# - Less predictable

# Best for:

# - Story writing
# - Brainstorming
# - Marketing
# - Creative content

# The course recommends using temperature 0 for most production applications.

# ---

# # Iterative Prompt Development

# Prompt engineering is an iterative process.

# Nobody writes the perfect prompt on the first attempt.

# Workflow:

# Write Prompt
# ↓
# Run Prompt
# ↓
# Inspect Output
# ↓
# Identify Problems
# ↓
# Improve Prompt
# ↓
# Repeat

# During the chair description example, the prompt was improved multiple times by adding:

# - Word limits
# - Audience specification
# - Product IDs
# - Dimension tables
# - HTML formatting

# Each improvement solved a specific problem and produced better output.

# ---

# # Summarization

# Large Language Models are excellent at summarizing text.

# Summaries can be controlled using:

# - Word limits
# - Sentence limits
# - Character limits

# Word and sentence limits generally work better than character limits.

# Example:

# Summarize this review in at most 30 words.

# ---

# ## Focused Summarization

# Instead of summarizing everything, focus on one aspect.

# Examples:

# Summarize only shipping information.

# Summarize only pricing feedback.

# Different departments can receive summaries relevant to their needs.

# ---

# ## Summarization vs Extraction

# Summarization provides an overall overview.

# Extraction returns only specific requested information.

# Example:

# Summarize:

# "The product arrived early, looked great, but was expensive."

# Summary:

# "The customer liked the product but thought it was expensive."

# Extraction:

# Shipping:
# Arrived one day early.

# Extraction is preferable when applications require specific facts instead of general summaries.

# ---

# # Inferring

# Inference means analyzing text instead of simply summarizing it.

# Common inference tasks include:

# - Sentiment Analysis
# - Emotion Detection
# - Topic Detection
# - Entity Extraction
# - Brand Detection
# - Classification

# ---

# ## Sentiment Analysis

# Example:

# Determine whether the sentiment is Positive or Negative.

# Output:

# Positive

# ---

# ## Emotion Detection

# Identify emotions present in a review.

# Example output:

# happy
# excited
# grateful

# ---

# ## Entity Extraction

# Extract information such as:

# - Product
# - Brand
# - Organization
# - Person

# Example:

# ```json
# {
#     "Item":"Lamp",
#     "Brand":"Lumina"
# }
# ```

# ---

# ## Unknown Values

# If information does not exist, instruct the model to return:

# unknown

# instead of inventing an answer.

# ---

# ## Multi-task Prompting

# Instead of making multiple API calls:

# Prompt 1:
# Sentiment

# Prompt 2:
# Brand

# Prompt 3:
# Emotion

# Combine everything into one prompt.

# Benefits:

# - Faster
# - Lower cost
# - Simpler backend logic

# ---

# ## Topic Extraction

# Determine the main topics discussed in a document.

# Example:

# NASA

# Government

# Engineering

# Healthcare

# Politics

# ---

# ## Topic Classification

# Rather than discovering topics, classify whether predefined topics are present.

# Example:

# NASA: 1

# Engineering: 0

# Healthcare: 0

# Useful for:

# - News categorization
# - Recommendation systems
# - Search indexing

# ---

# ## Zero-Shot Learning

# Zero-shot prompting means performing tasks without providing examples.

# Example:

# Determine whether this article discusses NASA.

# The model already understands the concept without additional training.

# ---

# # Transforming

# Transformation modifies existing text without changing its core meaning.

# Examples include:

# - Translation
# - Grammar correction
# - Tone adjustment
# - Format conversion

# ---

# ## Translation

# The model can:

# - Translate between many languages
# - Detect languages
# - Translate into multiple languages simultaneously
# - Translate into different writing styles

# Examples:

# English → Spanish

# English → French

# Formal Spanish

# Informal Spanish

# English Pirate

# ---

# ## Universal Translator

# One practical application demonstrated was an IT support system.

# Workflow:

# Receive customer messages in multiple languages

# Detect the language

# Translate to English

# Translate to another employee's language

# This enables multilingual customer support without human translators.

# ---

# ## Tone Transformation

# The model can rewrite text for different audiences.

# Example:

# Slang

# ↓

# Professional business letter

# ---

# ## Format Conversion

# Convert between formats.

# Examples:

# JSON

# ↓

# HTML Table

# Markdown

# ↓

# HTML

# Python Dictionary

# ↓

# Table

# ---

# ## Grammar and Spell Checking

# The model can:

# - Correct grammar
# - Fix spelling
# - Improve writing
# - Rewrite professionally
# - Follow writing styles such as APA

# The Redlines library was used to visually compare original and corrected text.

# ---

# # Expanding

# Expansion means generating longer text from shorter input.

# Examples:

# - Emails
# - Articles
# - Reports
# - Brainstorming

# The course demonstrated generating personalized customer service emails.

# Workflow:

# Review

# ↓

# Sentiment

# ↓

# Customized Email

# If sentiment is:

# Positive

# → Thank customer

# Negative

# → Apologize and suggest contacting customer support

# The email should:

# - Mention specific review details
# - Maintain a professional tone
# - Clearly identify itself as AI-generated

# This demonstrates combining multiple prompt engineering techniques into a complete application.

# ---

# # Temperature Revisited

# Temperature controls creativity.

# Temperature = 0

# Always produces nearly identical responses.

# Temperature = 0.7

# Produces different outputs for the same prompt.

# Lower temperatures are recommended for:

# - Customer support
# - JSON
# - Production

# Higher temperatures are useful for:

# - Story generation
# - Creative writing
# - Brainstorming

# ---

# # Chat Format

# The final lesson introduced the Chat Completion format.

# Instead of one prompt:

# Prompt

# ↓

# Response

# we send an entire conversation.

# Messages:

# System

# ↓

# User

# ↓

# Assistant

# ↓

# User

# ↓

# Assistant

# Each message contains:

# role

# content

# ---

# ## System Role

# Defines assistant behavior.

# Examples:

# You are a helpful teacher.

# You are Shakespeare.

# You are a customer support assistant.

# The system message controls personality, tone, behavior, and objectives.

# ---

# ## User Role

# Represents human input.

# ---

# ## Assistant Role

# Represents AI responses.

# These previous responses become part of the conversation history.

# ---

# # Context

# The model does not permanently remember previous conversations.

# Memory exists only because previous messages are included in the request.

# Conversation history is called Context.

# Without previous messages:

# User:
# What is my name?

# Model:

# I don't know.

# With previous conversation included:

# User:
# Hi, my name is Isa.

# Later:

# What is my name?

# The model correctly answers because the context was supplied.

# ---

# # Building Chatbots

# The course demonstrated building an OrderBot.

# The system prompt instructed the bot to:

# - Greet customers
# - Collect complete orders
# - Clarify sizes and toppings
# - Ask pickup or delivery
# - Collect address
# - Collect payment
# - Summarize the order

# Finally, another instruction generated a JSON summary containing:

# - Pizza
# - Size
# - Toppings
# - Drinks
# - Sides
# - Total price

# This demonstrated how prompts can control complete conversational workflows.

# ---

# # Hallucination

# Hallucination occurs when an LLM confidently generates incorrect or fabricated information.

# Prompt engineering cannot eliminate hallucinations completely, but it can significantly reduce them.

# Techniques include:

# - Writing clear prompts
# - Providing sufficient context
# - Using delimiters
# - Asking for "unknown" when information is missing
# - Requesting structured outputs
# - Breaking complex tasks into steps

# Good prompt design reduces ambiguity and therefore reduces hallucination.

# ---

# # Responsible AI

# Throughout the course, the instructors emphasized responsible AI usage.

# Large Language Models are powerful tools that should be used to:

# - Help users
# - Improve productivity
# - Solve problems
# - Create positive impact

# Developers should avoid building systems that:

# - Generate spam
# - Mislead users
# - Spread misinformation
# - Cause harm

# Transparency is also important. If content is AI-generated, users should be informed.

# ---

# # Overall Prompt Engineering Workflow

# Problem
# ↓
# Understand the Task
# ↓
# Write Clear Instructions
# ↓
# Use Delimiters
# ↓
# Specify Constraints
# ↓
# Choose Output Format
# ↓
# Give the Model Time to Think (if needed)
# ↓
# Run the Prompt
# ↓
# Inspect the Output
# ↓
# Improve the Prompt
# ↓
# Repeat Until Satisfied
# ↓
# Deploy in an Application

# ---

# # Key Takeaways

# - Prompt engineering allows one Large Language Model to perform many different tasks without retraining.
# - The two most important principles are writing clear instructions and giving the model time to think.
# - Good prompts specify the task, output format, constraints, audience, and style.
# - Delimiters help separate instructions from input data.
# - Structured outputs such as JSON are preferred for production systems.
# - Prompt engineering is an iterative process that improves through testing and refinement.
# - Large Language Models excel at summarization, inference, transformation, expansion, and conversational AI.
# - Multiple tasks can often be combined into a single prompt, reducing API calls and cost.
# - Temperature controls creativity and should be chosen based on the application's needs.
# - The Chat Completion format enables multi-turn conversations using system, user, and assistant messages.
# - Context is provided by including previous messages in each request; the model does not retain memory automatically.
# - Hallucinations can be reduced through better prompt design, sufficient context, and explicit instructions.
# - The best way to master prompt engineering is by building real applications, experimenting with prompts, and continuously refining them.
````
